```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here

fig, axa = plt.subplots(ncols = 3, sharex = True, sharey = True, figsize = (20,8))
#axa.figure(figsize=(7,5))
axa[0].plot(x,tan, 'blue',label = 'tanX')
axa[0].set_title('Grafik tanx')
axa[1].plot(x,sin,'red', label = 'sinX')
axa[1].set_title('Grafik sinx')
axa[2].plot(x,cos,'green', label = 'cosX')
axa[2].set_title('Grafik cosx')

lines_labels = [ax.get_legend_handles_labels() for ax in fig.axes]
lines, labels = [sum(sub, []) for sub in zip(*lines_labels)]

fig.legend(lines, labels,loc = 'upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(13,9))
ax.plot(x,y)
ax.set_title('Grafik CosX')
#axin1 = ax.inset_axes([-0.1, -0.1, 0.85, 0.15])
axin2 = ax.inset_axes([-8.8, -1.3, 4.1, 0.65], transform=ax.transData)
axin2.plot(x,y,x,y2,'red')
axin2.set_title('Grafik CosX^2')
axin3 = ax.inset_axes([5.4, -1.3, 4.1, 0.65], transform=ax.transData)
axin3.plot(x,y,x,y3,'green')
axin3.set_title('Grafik CosX^3')
axin4 = ax.inset_axes([5.4, 0.9, 4.1, 0.65], transform=ax.transData)
axin4.plot(x,y,x,y4,'black')
axin4.set_title('Grafik CosX^3')
axin5 = ax.inset_axes([-8.8, 0.9, 4.1, 0.65], transform=ax.transData)
axin5.plot(x,y,x,y2,'#cdd876')
axin5.set_title('Grafik CosX^4')
```
